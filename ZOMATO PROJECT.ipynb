```python
x=df_new.drop(['Price range'],axis=1)
y=df_new['Price range']
```

```python
#Scaling the x data
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
```

```python
for col in x:
    mms=MinMaxScaler()
    x[col]=mms.fit_transform(x[[col]])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score

import warnings
warnings.filterwarnings('ignore')
```

```python
knn=KNeighborsClassifier()
dtc=DecisionTreeClassifier()
svc=SVC()
mnb=MultinomialNB()

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)

Model=[knn,dtc,svc,mnb]

for m in Model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of',m, 'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

The above indicates that dtc model has the highest accuracy 

```python
#Cross Validation
```

```python
accuracy_score=cross_val_score(mnb,x,y,cv=5)
print(accuracy_score)  
print(accuracy_score.mean())
print(accuracy_score.std())
```

```python
accuracy_score=cross_val_score(dtc,x,y,cv=5)
print(accuracy_score) 
print(accuracy_score.mean())
print(accuracy_score.std())
```

```python
accuracy_score=cross_val_score(knn,x,y,cv=5)
print(accuracy_score) 
print(accuracy_score.mean())
print(accuracy_score.std())
```

```python
accuracy_score=cross_val_score(svc,x,y,cv=5)
print(accuracy_score) 
print(accuracy_score.mean())
print(accuracy_score.std())
```

The above indicates that 'dtc' has a higher accuracy score, thus could be chosen as the best for predicting test

```python
#Prediction
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x1_train,y1_train)

dtr.score(x1_train,y1_train)
preddtr=dtr.predict(x1_test)

print(preddtr)

dtr.score(x1_test,y1_test)
```

```python
knr=KNeighborsRegressor()
knr.fit(x1_train,y1_train)

knr.score(x1_train,y1_train)
predknr=knr.predict(x1_test)

print(predknr)

knr.score(x1_test,y1_test)
```

```python
svr=SVR(kernel='linear')
svr.fit(x1_train,y1_train)
svr.score(x1_train,y1_train)
pred_y=svr.predict(x1_test,)

svr=SVR(kernel='poly')
svr.fit(x1_train,y1_train)
svr.score(x1_train,y1_train)
pred_y=svr.predict(x1_test,)

svr=SVR(kernel='rbf')
svr.fit(x1_train,y1_train)
svr.score(x1_train,y1_train)
pred_y=svr.predict(x1_test,)
```

```python
kernellist=['linear','poly','rbf']
for i in kernellist:
    sv=SVR(kernel=i)
    sv.fit(x1_train,y1_train)
    print(sv.score(x1_train,y1_train))
```

From the above it is clear that KNeighborsRegressor is the best model as it has the highest score. However, using SVR, it is best to use "poly" as it has the highest score.
