```python
#removing outliers
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
#threshold for zscore values ---> above 3 are considered outliers

threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]   
```

```python
print('Old Dataframe',df.shape)
print('New Dataframe',df_new.shape)
print('Total_dropped_rows',df.shape[0] - df_new.shape[0])
```

The outliers in the data seems very insignificant.

```python
y=df_new.iloc[:,0]
x=df_new.iloc[:,1:]
```

```python
#Checking and removing skewness
```

```python
x.skew()
```

```python
from sklearn.preprocessing import power_transform,PowerTransformer
```

```python
x=power_transform(x,method='yeo-johnson')
x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=45)
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train,y_train)
```

```python
#predicting the 'Y' test values
```

```python
pred=lm.predict(x_test)
print('predict score:',pred)
print('actual score:',y_test)
```

```python
#Saving Model
```

```python
df=pd.DataFrame(pred)
df.to_csv('Baseball_project assignment.csv')
```
