```python
x=df_new.drop(['fraud_reported'],axis=1)
y=df_new['fraud_reported']
```

```python
#Scaling the x data
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
```

```python
for col in x:
    mms=MinMaxScaler()
    x[col]=mms.fit_transform(x[[col]])
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score

import warnings
warnings.filterwarnings('ignore')
```

```python
knn=KNeighborsClassifier()
dtc=DecisionTreeClassifier()
svc=SVC()
mnb=MultinomialNB()

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)

Model=[knn,dtc,svc,mnb]

for m in Model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy score of',m, 'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('\n')
```

The above indicates that SVC model has the highest accuracy 

```python
#Cross Validation
```

```python
accuracy_score=cross_val_score(mnb,x,y,cv=5)
print(accuracy_score)  
print(accuracy_score.mean())
print(accuracy_score.std())
```

```python
accuracy_score=cross_val_score(knn,x,y,cv=5)
print(accuracy_score)  
print(accuracy_score.mean())
print(accuracy_score.std())
```

```python
accuracy_score=cross_val_score(dtc,x,y,cv=5)
print(accuracy_score)  
print(accuracy_score.mean())
print(accuracy_score.std())
```

```python
accuracy_score=cross_val_score(svc,x,y,cv=5)
print(accuracy_score)  
print(accuracy_score.mean())
print(accuracy_score.std())
```

The above indicates that 'dtc' has a higher accuracy score, thus could be chosen as the best for predicting test
